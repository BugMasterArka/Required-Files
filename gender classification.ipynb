```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
```

```python
train_gen = ImageDataGenerator(rescale=1./255,
                              shear_range=0.2,
                              zoom_range=0.3,
                              horizontal_flip=True)
training_set = train_gen.flow_from_directory('../input/gender-classified-dataset-with-masked-face/GenderOcclusionData/Training',
                                            target_size=(64,64),
                                            batch_size=50,
                                            class_mode='binary')
```

```python
test_gen = ImageDataGenerator(rescale=1./255)
test_set = test_gen.flow_from_directory('../input/gender-classified-dataset-with-masked-face/GenderOcclusionData/Validation',
                                       target_size=(64,64),
                                       batch_size=50,
                                       class_mode='binary')
```

```python
cnn = tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=64,kernel_size=3,activation='relu',input_shape=(64,64,3)))
cnn.add(tf.keras.layers.Conv2D(filters=64,kernel_size=3,activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=128,kernel_size=3,activation='relu'))
cnn.add(tf.keras.layers.Conv2D(filters=128,kernel_size=3,activation='relu'))
cnn.add(tf.keras.layers.Conv2D(filters=128,kernel_size=3,activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units=512,activation='relu'))
cnn.add(tf.keras.layers.Dense(units=512,activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
cnn.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
cnn.fit(x=training_set,
        validation_data=test_set,
        epochs=15)
```
